```python
def convert_volume(value):
    if isinstance(value, str):
        value = value.replace(',', '').strip()  # Eliminăm virgulă și spații
        if value.endswith('K'):  # Dacă se termină cu 'K'
            return float(value[:-1]) * 1_000
        elif value.endswith('M'):  # Dacă se termină cu 'M'
            return float(value[:-1]) * 1_000_000
        else:  # Dacă nu are sufix
            return float(value)
    return value  # Returnăm valoarea originală dacă nu e string
```

```python
df['Vol.'] = df['Vol.'].apply(convert_volume)
```

```python
df['Vol.'] = df['Vol.'].fillna(method='ffill').fillna(method='bfill')
```

```python
# Calcularea Bollinger Bands
bb = ta.bbands(df['Close'], length=20, std=2)
df['BB_lower'] = bb['BBL_20_2.0']  # Banda inferioară
df['BB_mid'] = bb['BBM_20_2.0']    # Banda mediană
df['BB_upper'] = bb['BBU_20_2.0']  # Banda superioară
```

```python
df['OBV'] = ta.obv(df['Close'], df['Vol.'])
```

```python
# Calculează PSAR folosind pandas_ta
psar_values = ta.psar(df['High'], df['Low'], df['Close'])
```

```python
stochastic = ta.stoch(high=df['High'], low=df['Low'], close=df['Close'], k=14, d=3)
```
