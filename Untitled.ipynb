```python
interval = 14
```

```python
df['-DM'] = df['Low'].shift(1) - df['Low']
df['+DM'] = df['High'] - df['High'].shift(1)
df['+DM'] = np.where((df['+DM'] > df['-DM']) & (df['+DM'] > 0), df['+DM'], 0.0)
df['-DM'] = np.where((df['-DM'] > df['+DM']) & (df['-DM'] > 0), df['-DM'], 0.0)
```

```python
df['TR_TMP1'] = df['High'] - df['Low']
df['TR_TMP2'] = np.abs(df['High'] - df['Price'].shift(1))
df['TR_TMP3'] = np.abs(df['Low'] - df['Price'].shift(1))
df['TR'] = df[['TR_TMP1', 'TR_TMP2', 'TR_TMP3']].max(axis=1)
```

```python
df = df.drop(['TR_TMP3'],axis=1)
```

```python
for i in range(interval-1, len(df['Price'])):
    if(i == interval - 1):
        df.loc[i, 'TR_smt'] = df.loc[0:interval, 'TR'].mean()
        df.loc[i, '+DM_smt'] = df.loc[0:interval, '+DM'].mean()
        df.loc[i, '-DM_smt'] = df.loc[0:interval, '-DM'].mean()
    else:
        df.loc[i, 'TR_smt'] = ((df.loc[i-1, 'TR_smt'] * (13)) + df.loc[i, 'TR']) / 14
        df.loc[i, '+DM_smt'] = ((df.loc[i-1, '+DM_smt'] * (13)) + df.loc[i, '+DM']) / 14
        df.loc[i, '-DM_smt'] = ((df.loc[i-1, '-DM_smt'] * (13)) + df.loc[i, '-DM']) / 14
```

```python
df['+DI'] = (df['+DM_smt'] / df['TR_smt']) * 100
df['-DI'] = (df['-DM_smt'] / df['TR_smt']) * 100
```

```python
df['DX'] = (abs(df['+DI'] - df['-DI']) / abs(df['+DI'] + df['-DI'])) * 100
```

```python
for i in range(2 * (interval - 1), len(df['Price'])):
    if(i == 2 * (interval - 1)):
        df.loc[i, 'ADX'] = df.loc[(interval - 1): 2 * (interval - 1), 'DX'].mean()
    else:
        df.loc[i, 'ADX'] = ((df.loc[i-1, 'ADX'] * 13) + df.loc[i, 'DX']) / 14
```

```python
fig, axis = plt.subplots(2, 1)
axis[0].plot(df['Price'])
axis[1].plot(df['ADX'], label='ADX')
axis[1].plot(df['+DI'], label='+DI')
axis[1].plot(df['-DI'], label='-DI')
plt.legend()
plt.show()
```

```python
# Create subplot with 2 rows and shared x-axis
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, vertical_spacing=0.02)

# Add candlestick chart
fig.add_trace(go.Candlestick(
        x=df['Date'],
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Price'],
        name='Candlestick'
    )
        ,row=1, col=1)


# ADX, +DI, and -DI line charts
fig.add_trace(go.Scatter(x=df['Date'], y=df['ADX'], mode='lines', name='ADX', line=dict(color='blue')),
              row=2, col=1)
fig.add_trace(go.Scatter(x=df['Date'], y=df['+DI'], mode='lines', name='+DI', line=dict(color='green'), opacity=0.2),
              row=2, col=1)
fig.add_trace(go.Scatter(x=df['Date'], y=df['-DI'], mode='lines', name='-DI', line=dict(color='red'), opacity=0.2),
              row=2, col=1)

# Update layout
fig.update_layout(yaxis1_title='Price ', yaxis2_title='+DI / -DI / ADX',
                  showlegend=True, height=1000, xaxis_rangeslider_visible=False)

# Remove x-axis tick labels for the first chart
fig.update_xaxes(showticklabels=False, row=1, col=1)

fig.show()
```

```python
# Calculul RSI
def calculate_rsi(data, window=14):
    delta = data['Price'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df['RSI'] = calculate_rsi(df, window=14)
```

```python

# Crearea graficului doar pentru RSI
fig = make_subplots(rows=1, cols=1)

# AdÄƒugare RSI
fig.add_trace(go.Scatter(
    x=df.index, 
    y=df['RSI'], 
    mode='lines', 
    name='RSI(14)',
    line=dict(color='blue', width=2)
), row=1, col=1)

# AdÄƒugare linii supracumpÄƒrare È™i supravÃ¢nzare
fig.add_hline(y=70, line=dict(color='green', dash='dash'), row=1, col=1)
fig.add_hline(y=30, line=dict(color='red', dash='dash'), row=1, col=1)

fig.add_trace(go.Scatter(
    x=df.index, 
    y=[70] * len(df), 
    fill='tonexty', 
    mode='lines', 
    line=dict(color='red'), 
    showlegend=False
), row=1, col=1)




# Layout
fig.update_layout(
    title='RSI(14)',
    xaxis_title='Date',
    yaxis_title='RSI',
    template='plotly_white'
)

# AfiÈ™eazÄƒ graficul
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Date de exemplu
x = np.linspace(0, 10, 100)
y = np.sin(x) - 0.5  # date care au valori pozitive È™i negative

# Crearea figurii È™i axei
fig, ax = plt.subplots()

# SeparÄƒm porÈ›iunile de deasupra È™i sub axa 0
ax.plot(x, np.where(y >= 0, y, np.nan), color="red", label="Above 0")  # Doar valorile deasupra 0 Ã®n roÈ™u
ax.plot(x, np.where(y < 0, y, np.nan), color="green", label="Below 0")  # Doar valorile sub 0 Ã®n verde

# AdÄƒugÄƒm linia de axÄƒ 0 pentru referinÈ›Äƒ
ax.axhline(0, color='black', linewidth=0.5)

# SetÄƒri adiÈ›ionale pentru claritate
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.legend()
plt.title("Line chart cu zone deasupra È™i dedesubt de 0 colorate diferit")

# AfiÈ™Äƒm graficul
plt.show()
```
